# Data clean up – correcting the data types

__1. Which all variables need to be brought to numeric types?__

__2.Price variable- remove '$' sign and convert to float__

__3. Installs – remove ‘,’ and ‘+’ sign, convert to integer__

4.Convert all other identified columns to numeric

```python
#1
df2.head()
```

```python
df2_new['Reviews'] = df2_new['Reviews'].astype(int)
df2_new['Size'] = df2_new['Size'].astype(int)
# Additional conversion
df2_new['Last Updated'] = df2_new['Last Updated'].astype('datetime64[ns]')
```

```python
#2
df2_new['Price'] = df2_new['Price'].str.replace('$', '')
```

```python
df2_new.head(2)
```

```python
df2_new['Price'] = df2_new['Price'].astype(float)
```

```python
df2_new.info()
```

```python
#3
df2_new['Installs'] = df2_new['Installs'].str.replace('+','')
```

```python
df2_new['Installs'] = df2_new['Installs'].str.replace(',', '')
```

```python
df2_new['Installs'] = df2_new['Installs'].astype(int)
```

```python
df2_new.head(2)
```

```python
df2_new.info()
```
