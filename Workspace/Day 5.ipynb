# Different possible combination to merge the data

### Combining Datasets : Merge & Join

```python
# One to One Join - column wise concatination

df1= pd.DataFrame({
    'employee':['Bob','Jake','Lisa','Sue'],
    'group' : ['Accounting','Engineering','Engineering','HR']
})
df2= pd.DataFrame({
    'employee':['Lisa','Bob','Jake','Sue'],
    'hire_date' : [2004,2008,2012,2014]
})


#print both data frame

print(df1)
print()
print(df2)
#Merge
df3=pd.merge(df1,df2)
df3

```

```python
# One to One Join Note:
#For merge operation to work there should be atleast **one common column in both the data set**

#
```

```python

# MANY TO ONE JOINS - One of the columns should have duplicate entries

df4=pd.DataFrame({
    'group':['Accounting','Engineering','HR'],
    'supervisor':['Carly','Guido','Steve']
})
print(df3)
print()
print(df4)

#Merge
pd.merge(df3,df4)
```

```python
#Note

#Goup is common in both df3 and df4 data frame
#One supervisor can be assigned to many employees/group

```

```python
# MANY TO MANY JOIN - very defined 

df5 = pd.DataFrame({
    'group' : ['Accounting','Accounting','Engineering','Engineering','HR','HR'],
    'skills' : ['maths', 'spreadsheets','coding','linux','spreadsheets','organisation']
})
print(df1)
print()
print(df5)

#merge
pd.merge(df1,df5)
```

# Join

```python
# Inner And Outer Join
df6 = pd.DataFrame({
    'name':['Peter','Paul','Mary'],
    'food': ['fish','beans','bread']
})
df7 = pd.DataFrame({
    'name':['Mary','Joseph'],
    'drink':['wine','beer']
})

print(df6)
print()
print(df7)
#by default this takes Inner Join
pd.merge(df6,df7)
```

```python
# Inner And Outer Join
df6 = pd.DataFrame({
    'name':['Peter','Paul','Mary'],
    'food': ['fish','beans','bread']
})
df7 = pd.DataFrame({
    'name':['Mary','Joseph'],
    'drink':['wine','beer']
})

print(df6)
print()
print(df7)
#by default this takes Inner Join
pd.merge(df6,df7, how = 'inner')
```

```python
#outer join
print(df6)
print()
print(df7)

pd.merge(df6, df7, how='outer')
```

# GroupBy : Split,Apply,Combine


#### (very impt) 

key - catagorical , data - continious

key - a,b,c,a,b,c
data - 1,2,3,4,5,6

Split - A => key: a,a
             data:1,4
Split - B => key: b,b
             data:2,5
Split - C => key: c,c
             data:3,6

(Apply=sum)
Apply - A => key: a
             data:5
Apply - B => key: b
             data:7
Apply - C => key: c
             data:9
             
Combine - key: a, b, c
          data:5, 7, 9

```python
df = pd.read_csv("https://raw.githubusercontent.com/ingledarshan/BoardInfinity/main/student_records.csv")
df
```

```python
#Filtering

df['ResearchScore']>50
```

```python
#Filtering - from my df where do my RS is > 50

df[(df['ResearchScore']>50) & (df['Recommend']=='Yes')]
```

```python
df
```

```python
#How to apply a fn over some column of the dataframe?
#Assume RS is very poor. Now i want to add 5 marks for each student
# if a student has less than <= 35 and >= 30 marks in Project Score,
#add a grace of 5 marks to their score
#Assuming that 35 is the passing marks

#Creating fn
#defining grace
def grace(x):
    if x>= 30 and x<= 35:
        x = x+5
        return x
    else:
        return x
    

```

```python
df['ProjectScore']=df['ProjectScore'].apply(grace)
df
```

```python
# If the student has >=35 marks in PS , Recommend them with 'Yes'

#using lambda fn
df['Recommend'] = df['ProjectScore'].apply(lambda x: 'Yes' if x>=35 else 'No')
df
```

```python
#Traditional way
# If the student has >=65 marks in RS , Recommend them with 'Yes

def rec_fun(x):
    if x>=65:
        return 'Yes'
    else:
        return 'No'
```

```python
df['Recommend']=df['ResearchScore'].apply(rec_fun)
df
```
